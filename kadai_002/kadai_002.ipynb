```python
import pandas as pd
```

```python
data = {'日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
        '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
        '売上': [100, 200, 150, 300, 250],
        '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店']}

df = pd.DataFrame(data)
df
```

```python
df['平均売上'] = df['売上'].mean()
df
```

```python
def rank(sales, mean):
    if sales >= mean + 50:
        return 'A'
    elif sales >= mean:
        return 'B'
    else:
        return 'C'
```

```python
df['業績ランク'] = df['売上'].apply(rank, args=(df['平均売上'].mean(),))
df
```

```python
writer = pd.ExcelWriter('業績.xlsx')
df.to_excel(writer, sheet_name='Sheet1', index=False)
writer.close()
```
