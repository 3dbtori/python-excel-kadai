```python
import pandas as pd
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.styles import PatternFill
```

```python
file_names = [r'課題3/2022_年間売上表.xlsx', r'課題3/2023_年間売上表.xlsx']
df = pd.DataFrame()

for file_name in file_names:
    df_temp = pd.read_excel(file_name)
    df = pd.concat([df, df_temp])
    
df = pd.DataFrame(df.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum())
df
```

```python
wb = openpyxl.Workbook()
ws = wb.active

for r in dataframe_to_rows(df, index=False, header=True):
    ws.append(r)

for rows in ws['A1:C1']:
    for cell in rows:
        cell.fill = PatternFill(patternType='solid', fgColor='F2F2F2')

wb.save('売上集計表.xlsx')
```
