```python
import openpyxl
from datetime import datetime
from zoneinfo import ZoneInfo

wb = openpyxl.Workbook()
ws = wb.active

title = '請求書'
ws['B2'] = title

cus_info = ['株式会社ABC', '〒101-0022 東京都千代田区神田練塀町300',
            'TEL:03-1234-5678 FAX:03-1234-5678', '担当者名:鈴木一郎 様']

for i, info in enumerate(cus_info):
    ws.cell(row=i+4, column=2, value=info)

today = datetime.now(ZoneInfo('Asia/Tokyo'))

y = today.year
m = today.month
d = today.day
data = f'{y}/{m:0>2}/{d:02}'

str_nom = '0001'
for i, row in enumerate([('No.', str_nom), ('日付', data)]):
    v1, v2 = row
    ws.cell(row=4+i, column=6, value=v1)
    ws.cell(row=4+i, column=7, value=v2)

header = ['商品名', '数量','単価', '金額']
data_dict = {'商品A': (2, 10000), '商品B': (1, 15000)}

now_row = ws.max_row

for i , data in enumerate(header):
    ws.cell(row=now_row+3, column=2+i, value=data)

sum_total_money = 0
for i, data in enumerate(data_dict.items()):
    k, v = data
    num, money = v
    total_money = money * num
    sum_total_money += total_money
    ws.cell(row=now_row+4+i, column=2, value=k)
    ws.cell(row=now_row+4+i, column=3, value=num)
    ws.cell(row=now_row+4+i, column=4, value=money)
    ws.cell(row=now_row+4+i, column=5, value=total_money)
ws.cell(row=now_row+5+i, column=5, value=sum_total_money)

now_row = ws.max_row
sum_dict = {'合計': sum_total_money, '消費税': sum_total_money * 0.1, '税込合計': sum_total_money * 1.1}
for i, data in enumerate(sum_dict.items()):
    k, v = data
    ws.cell(row=now_row+2+i, column=2, value=k)
    ws.cell(row=now_row+2+i, column=5, value=v)

wb.save(f'{title}.xlsx')
```
